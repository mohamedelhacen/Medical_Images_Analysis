```python
for i in range(9):
    rand = random.randint(0, len(all_files)-1)
    img = pydicom.read_file(all_files[rand]).pixel_array
    plt.subplot(3, 3, i+1)
    plt.imshow(img, cmap='gray')
    plt.title(rand)
    plt.axis('off')
plt.show()
```
